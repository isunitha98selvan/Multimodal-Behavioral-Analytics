```python
x_df = x_df.drop(x_df.columns[0], axis=1)
y_df = y_df.drop(y_df.columns[0], axis=1)
y_df = y_df.drop(y_df.columns[-1],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

clf = SVR(kernel='rbf',C=1.0, epsilon=0.2)
length = y_train.shape[1]
for i in range(length):
    y  = y_train.iloc[:, [i]].values.ravel()
    clf.fit(X_train, y)
    y_pred = clf.predict(X_test)
    y_true = y_test.iloc[:,[i]].values.ravel()
#     print("Predicted: ",y_pred)
#     print("True values: ", y_true)
    print(clf.score(X_test, y_true))
#     print(clf.accuracy_score(y_true,y_pred))
#     plt.scatter(X_test.iloc[:,0], y_true, color = 'purple')
#     plt.plot(X_test, clf.predict(X_test), color = 'green')
#     plt.title('Support Vector Regression Model')
#     plt.xlabel('x values')
#     plt.ylabel('Score')
#     plt.show()
#     break
```
